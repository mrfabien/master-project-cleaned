```python
import pandas as pd
import numpy as np
```

```python
# order each tracks by step
path_eu = r'C:\Users\fabau\OneDrive\Documents\GitHub\master-project-cleaned\pre_processing\tracks\ALL_TRACKS\tracks_1h_EU/'
path_non_eu = r'C:\Users\fabau\OneDrive\Documents\GitHub\master-project-cleaned\pre_processing\tracks\ALL_TRACKS\tracks_1h_non_EU/'
paths = [path_eu, path_non_eu]
for path in paths:
    for i in range(1, 97):
        storm = pd.read_csv(path + f'storm_{i}.csv', index_col=0)
        # order the rows by the column step
        storm = storm.sort_values(by='step')
        # save the ordered dataframe
        storm.to_csv(path + f'storm_{i}.csv')
```

```python
# for non_eu tracks, create a new folder with all the tracks that are before coordinate 26° West, but not above 180° West
path_non_eu_before_16 = r'C:\Users\fabau\OneDrive\Documents\GitHub\master-project-cleaned\pre_processing\tracks\ALL_TRACKS\tracks_1h_before_EU/'
for i in range(1, 97):
    storm = pd.read_csv(path_non_eu + f'storm_{i}.csv', index_col=0)
    storm_temp = []
    for j in range(0,len(storm)):
        lon = storm.iloc[j]['lon_west'] + 4
        # check if the track is before 26° West
        if lon < 16 or lon >= 180:
            storm_temp.append(storm.iloc[j])
        else:
            continue
    storm_temp = pd.DataFrame(storm_temp)
    storm_temp.to_csv(path_non_eu_before_16 + f'storm_{i}.csv')
```

```python
# check if the tracks are continuous

count = 0
for i in range(1, 97):
    storm = pd.read_csv(path_non_eu_before_16 + f'storm_{i}.csv', index_col=0)
    # check if the steps are continuous
    if len(storm) == 0:
        print(f'storm_{i} is empty')
    else:
        steps = storm['step'].values
        diff = np.diff(steps) 
        if diff.std() > 1:
            print(f'storm_{i} is not continuous')
            count += 1
        else:
              print(f'storm_{i} is continuous')

print(f'Number of non-continuous tracks: {count}')
```

```python
# remove the tracks that are after the discontinuity
path_continuous = r'C:\Users\fabau\OneDrive\Documents\GitHub\master-project-cleaned\pre_processing\tracks\ALL_TRACKS\tracks_1h_before_EU_continuous/'
for i in range(1, 97):
    storm = pd.read_csv(path_non_eu_before_16 + f'storm_{i}.csv', index_col=0)
    # check if the steps are continuous
    if len(storm) == 0:
        continue
    else:
        steps = storm['step'].values
        diff = np.diff(steps) 
        if diff.std() > 1:
            # find the index of the discontinuity
            idx = np.where(diff > 1)[0][0]
            # remove the rows after the discontinuity
            storm = storm.iloc[:idx+1]
            storm.to_csv(path_non_eu_before_16 + f'storm_{i}.csv')
```
