```python
# create a list with the levels, if a var is SL, then the level is 0

var_pl = ['relative_humidity', 'specific_rain_water_content', 'vertical_velocity']
levels = pd.read_csv('data/levels.csv')
geop_levels = [500, 1000]

var_list_levels = []

for var in name_of_variable['variables']:
    if var in var_pl:
        for level in levels['levels']:
            if level == 0:
                continue
            else:
                var_list_levels.append(var + '_' + str(level))
    elif var == 'geopotential':
        for geop_level in geop_levels:
            var_list_levels.append(var + '_' + str(geop_level))
    else:
        var_list_levels.append(var)

stats = ['max', 'min', 'mean', 'std']

var_list_levels_stats = []

for var in name_of_variable['variables']:
    for stat in stats:
        if var in var_pl:
            for level in levels['levels']:
                if level == 0:
                    continue
                else:
                    var_list_levels_stats.append(var + '_' + str(level)+ '_' + stat)
        elif var == 'geopotential':
            for geop_level in geop_levels:
                var_list_levels_stats.append(var + '_' + str(geop_level)+ '_' + stat)
        else:
            var_list_levels_stats.append(var+ '_' + stat)

# Name the column variables

var_list_levels = pd.DataFrame(var_list_levels)
var_list_levels.columns = ['variables']

var_list_levels_stats = pd.DataFrame(var_list_levels_stats)
var_list_levels_stats.columns = ['variables']

var_list_levels.to_csv('data/variable_list_levels.csv', index=False)
var_list_levels_stats.to_csv('data/variable_list_levels_stats.csv', index=False)
```
